```python
import os, sys
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import subprocess
```

```python
root =  '/usr/local/serenceslab/maggie/shapeDim/'

# for ss in [1]:
    
for ss in np.arange(2, 11):
    
    subj = 'S%02d'%ss

    func_path = os.path.join(root, 'DataPreproc', subj)
    runs_list = os.path.join(func_path, 'runs.list')
    print(runs_list)

    with open(runs_list, 'r') as f:
        lines = f.read()

    lines = lines.split('\n')
    lines_rep = [l for l in lines if 'rep' in l]
    lines_rep

    new_folder = os.path.join(func_path, '%s_data_repeat_task'%subj)
    if not os.path.exists(new_folder):
        os.makedirs(new_folder)

    for l in lines_rep:

        sess = l.split('_')[0]
        run = l.split('_')[1].split(' ')[0]

        full_filename = os.path.join(func_path, '%s_REG_MC_DET_%s.nii.gz'%(sess, run))
        print(full_filename)

        new_filename = os.path.join(new_folder, '%s_REG_MC_DET_%s.nii.gz'%(sess, run))
        cmd = 'cp %s %s'%(full_filename, new_filename)
        print(cmd)

        err = subprocess.run(cmd, shell=True)


    os.chdir(func_path)

    folder_to_zip = '%s_data_repeat_task/*'%(subj)

    file_to_zip = '%s_data_repeat_task.zip'%(subj)

    cmd = 'zip %s %s'%(file_to_zip, folder_to_zip)
    print(cmd)

    err = subprocess.run(cmd, shell=True)

    print(err)
```
