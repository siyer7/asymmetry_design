```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import os, sys
import pandas as pd
import PIL.Image
import scipy.io

root = '/usr/local/serenceslab/maggie/shapeDim/'

sys.path.append(os.path.join(root, 'Analysis'))
from code_utils import file_utils, data_utils, plot_utils
```

```python
subjects = np.arange(1, 11)
n_subjects = len(subjects)

roi_names = ['V1', 'V2', 'V3', 'V3AB', 'hV4', 'LO1', 'LO2', 'IPS']
n_rois = len(roi_names)

final_counts_all = np.zeros((n_subjects, n_rois))

make_time_resolved=False
use_bigIPS = True; 
concat_IPS = True;


for si, ss in enumerate(subjects):


    print(ss)
    
    main_data, _, main_labels, roi_names_check = data_utils.load_main_task_data(ss, make_time_resolved, \
                                                                                 use_bigIPS, concat_IPS)

    print(roi_names_check)
    n_vox = np.array([main_data[ii].shape[1] for ii in range(n_rois)])
    final_counts_all[si,:] = n_vox

```

```python
vox_df = pd.DataFrame(final_counts_all.astype(int).T, index=roi_names, \
                      columns=['S%02d'%ss for ss in subjects])

figfolder = os.path.join(root, 'Analysis','figures')
vox_counts_filename = os.path.join(figfolder, 'voxel_counts_final.csv')
print(vox_counts_filename)
vox_df.to_csv(vox_counts_filename)

vox_df
```
