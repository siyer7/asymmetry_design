grab channels to resort

```python
merger_chan_dir = glob.glob(f'../../data/units/mergers/2025{patient}/*.png')
merger_chans = set([os.path.basename(fname).split('_')[0][1:] for fname in merger_chan_dir])
print(f'merger channels: {merger_chans}')

questionable_chan_dir = glob.glob(f'../../data/units/questionables/2025{patient}/*.png')
questionable_chans = set([os.path.basename(fname).split('_')[0][1:] for fname in questionable_chan_dir])
print(f'questionable channels: {questionable_chans}')

chans_to_resort = merger_chans.union(questionable_chans)
print(f'channels to resort: {chans_to_resort}')
```
